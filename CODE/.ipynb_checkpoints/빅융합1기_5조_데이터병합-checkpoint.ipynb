```python
sales_19 = pd.read_csv('DATA/상권_추정매출_2019.csv',low_memory=False)
sales_20 = pd.read_csv('DATA/상권_추정매출_2020.csv',low_memory=False)
pop      = pd.read_csv('DATA/상권_생활인구.csv',low_memory=False)
worker   = pd.read_csv('DATA/상권_직장인구.csv',low_memory=False)
```
